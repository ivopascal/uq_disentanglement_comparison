```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dataset['marker'] = ['s'] * 4 + ['p'] * 4 + ['P'] * 4  + ['*'] * 4
```

```python
for index, row in df_dataset.iterrows():
    plt.scatter([row["GL EU"]], [row["GL AU"]], label="Gaussian Logits", marker=row['marker'], color='blue')
    plt.scatter([row["IT EU"]], [row["IT AU"]], label="Information Theoretic", marker=row['marker'], color='orange')
plt.scatter([1], [0], s=100, label="Ideal Behaviour", color='green')
plt.vlines([0], ymin=-1, ymax=1, color='black', linestyles='--')
plt.hlines([0], xmin=-1, xmax=1, color='black', linestyles='--')
plt.xlabel("Epistemic Uncertainty")
plt.ylabel("Aleatoric Uncertainty")
plt.title("Correlations for Decreasing Dataset")
# plt.legend()
plt.show()
```

```python
species = ("Adelie", "Chinstrap", "Gentoo", "Emperor")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

uq_names = ("MC-Dropout", "MC-DropConnect", "Flipout", "Deep Ensembles")
uq_names_abbreviated = ("Dropout", "DropConnect", "Flipout", "Deep Ens.")



uq_means = {
    "GL AU": [df_dataset[df_dataset.iloc[:, 1] == uq_name]["GL AU"].mean() for uq_name in uq_names] ,
    "IT AU": [df_dataset[df_dataset.iloc[:, 1] == uq_name]["IT AU"].mean() for uq_name in uq_names],
    "GL EU": [df_dataset[df_dataset.iloc[:, 1] == uq_name]["GL EU"].mean() for uq_name in uq_names],
    "IT EU": [df_dataset[df_dataset.iloc[:, 1] == uq_name]["IT EU"].mean() for uq_name in uq_names],
}
```

```python
import numpy as np
```

```python
for attribute, measurement in uq_means.items():
    print(attribute)
    print(measurement)
```

```python
import matplotlib
matplotlib.rc('ytick', labelsize=14) 


n_datasets = 4
x = np.arange(n_datasets)
width = 0.4
multiplier_eu = -1
multiplier_au = -1

fig, axes = plt.subplots(2, 1, layout='constrained', figsize=(6, 5.5), sharex=True, sharey=True)

for attribute, measurement in uq_means.items():
    print(attribute)
    if "AU" in attribute:
        hatch = ""
        extra_offset = 0.02
        ax = axes[0]
        multiplier_au += 1
        multiplier = multiplier_au

    elif "EU" in attribute:
        hatch = ""
        extra_offset = -0.02
        ax = axes[1]
        multiplier_eu += 1
        multiplier = multiplier_eu

    extra_offset = -0.2

    offset = width * multiplier

    if "GL" in attribute:
        color = "C0"
    elif "IT" in attribute:
        color = "C1"
    rects = ax.bar(x + offset - extra_offset, measurement, width, label=attribute, hatch=hatch, color=color)
    # ax.bar_label(rects, padding=3)

# Add some text for labels, title and custom x-axis tick labels, etc.
# ax.set_ylabel('Length (mm)')
# ax.set_title('Uncertainty correlations - Changing dataset size')
axes[0].set_xticks(x + width, uq_names, fontsize=14)
# axes[0].set_title("Aleatoric Correlations", fontsize=14)
# axes[1].set_title("Epistemic Correlations", fontsize=14)
# axes[0].legend(loc='upper right')
axes[1].set_xticks(x + width, uq_names, fontsize=14)
axes[1].legend(loc='lower right', labels=["Gaussian Logits", "Information Theoretic"] , fontsize=14)
# ax.set_ylim(0, 250)
axes[0].set_ylabel("Correlation coefficient", fontsize=14)


axes[0].set_xticks(x + width, uq_names_abbreviated, fontsize=14)
# axes[0].set_title("Aleatoric Correlations", fontsize=23)
# axes[1].set_title("Epistemic Correlations", fontsize=23)
# axes[0].legend(loc='upper right')
axes[1].set_xticks(x + width, uq_names_abbreviated, fontsize=14)
axes[1].legend(loc='lower right', labels=["Gaussian Logits", "Information Theoretic"] , fontsize=14)

axes[0].hlines(0, xmin=0, xmax=multiplier_au * 4 + extra_offset, color='black', linestyle='dashed', label="_dashed_line")
axes[1].hlines(0, xmin=0, xmax=multiplier_eu * 4 + extra_offset, color='black', linestyle='dashed', label="_dashed_line")


# ax.set_ylim(0, 250)
# axes[0].hlines(0, xmin=0, xmax=multiplier_au * 4 + extra_offset, color='black', linestyle='dashed')
# axes[1].hlines(0, xmin=0, xmax=multiplier_eu * 4 + extra_offset, color='black', linestyle='dashed')
axes[0].set_ylabel("Aleatoric", fontsize=18)
axes[1].set_ylabel("Epistemic", fontsize=18)
axes[0].yaxis.set_label_position("right")
axes[1].yaxis.set_label_position("right")

fig.supylabel("Correlation coefficient", fontsize=18)

plt.savefig("dataset_size_correlation_coefficients_vert.pdf")

plt.show()
```
